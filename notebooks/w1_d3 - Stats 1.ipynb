# Probs del dado

```python
def par(x):
    return x%2==0
```

```python
sample_space=[1,2,3,4,5,6]

caras=len(sample_space)

pares=len([e for e in sample_space if par(e)])

pares_prob=pares/caras

pares_prob
```

# Distribuciones Discretas

### binomial

```python
# https://es.wikipedia.org/wiki/Distribuci%C3%B3n_bin%C3%B3mica

from scipy.stats import binom
import matplotlib.pyplot as plt

# binom.pmf(k, n, p)  ; k=nº de exitos, n=nº de repeticiones, p=prob del evento

binom.pmf(3, 10, 0.5)   # ¿Cual es la prob de que salgan 3 caras si tiro 10 veces la moneda?
```

```python
binom.pmf(5, 20, 1/6)  # ¿Cual es la prob de que salgan 5 doses si tiro 20 veces el dado?
```

```python
x=[i for i in range(20)]

y=[binom.pmf(x, 20, 0.5) for x in x]

plt.plot(x, y)
sum(y)
```

```python
x=[i for i in range(20)]

y=[binom.cdf(x, 20, 0.8) for x in x]

plt.plot(x, y);
```

```python
# https://es.wikipedia.org/wiki/Momento_central

n=10
p=0.5


media, var, sime, kur = binom.stats(n, p, moments='mvsk')
```

### uniforme discreta

```python
# https://es.wikipedia.org/wiki/Distribuci%C3%B3n_uniforme_discreta
```

```python
from scipy.stats import randint

randint.pmf([0,6,34,9,10], 0, 8)  # (elementos, intervalo)
```

```python
x=[i for i in range(-5, 20)]

y=[randint.pmf(x, 0, 8) for x in x]

plt.scatter(x, y);
```

```python
x=[i for i in range(-5, 20)]

y=[randint.cdf(x, 0, 8) for x in x]

plt.scatter(x, y);
```
