# Modelo Maher (estimacion maximo verosimil)

```python
def medias(params, Equipo, Oponente):
    
    lamb=np.exp(params['gamma']+params['alfa_'+Equipo]+params['beta_'+Oponente])
    mu=np.exp(params['alfa_'+Oponente]+params['beta_'+Equipo])
    
    return [lamb, mu]
```

```python
def encuentro(params, Equipo, Oponente, g_max):
    avgs=medias(params, Equipo, Oponente)
    
    pred=[[poisson.pmf(i, avg) for i in range(0, g_max)] for avg in avgs]
    
    matriz=np.outer(np.array(pred[0]), np.array(pred[1]))
    
    return matriz
```

```python
def loglike(x, y, alfai, betai, alfaj, betaj, gamma):
    lamb=np.exp(gamma+alfai+betaj)
    mu=np.exp(alfaj+betai)
    
    l=-lamb+(x*np.log(lamb))-mu+(y*np.log(mu))
    return l
```

```python


def parametros(datos,opciones={'disp':True,'maxiter':100},  # Funcion para obtener alfas, betas, ro y gamma, stats y maximas iteraciones
                 metodo='trust-constr'):    # metodo Nelder-Mead, Powell, CG, BFGS, Newton-CG, L-BFGS-B, TNC, COBYLA, SLSQP, trust-constr, dogleg, trust-ncg, trust-exact, trust-krylov (tambien custom)
                 

    equipos=np.sort(datos['Equipo'].unique())      # array equipos locales
    oponentes=np.sort(datos['Oponente'].unique())  # array equipos visitantes
     
    n_equipos=len(equipos)      # numero de equipos
    
    # inicializacion aleatoria de los parametros
    val_inicial=np.concatenate((np.random.uniform(0,1,(n_equipos)),  # alfa ataque
                                np.random.uniform(0,-1,(n_equipos)), # beta defensa
                                np.array([1.0])))                    # gamma factor casa
  

    def menos_loglike(params):  # Funcion menos log-verosimilitud, funcion a minimizar
        alfa_coefs = dict(zip(equipos, params[:n_equipos]))
        beta_coefs = dict(zip(equipos, params[n_equipos:(2*n_equipos)]))
        gamma = params[-1:]
        log_like = [loglike(n.GolesEq, n.GolesOp, alfa_coefs[n.Equipo], beta_coefs[n.Equipo],
                     alfa_coefs[n.Oponente], beta_coefs[n.Oponente], gamma) for n in datos.itertuples()] # seleccion en filas
        
        return -sum(log_like) 
        
    print ('Optimizacion %s...' % metodo)    
    salida=minimize(menos_loglike, val_inicial, options=opciones, method=metodo) # minimizacion

    return dict(zip(["alfa_"+equipo for equipo in equipos]+   # devuelve como diccionario 
                    ["beta_"+equipo for equipo in equipos]+['gamma'], salida.x))
```

```python
p=parametros(datos)
p
```

```python
M=encuentro(p, 'Valencia', 'Real Madrid', 10)
```
